```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df_jeopardy = pd.read_csv('jeopardy.csv')
df_jeopardy.columns = df_jeopardy.columns.str.lower().str.replace(' ', '_')
df_jeopardy.columns = df_jeopardy.columns.str.lstrip('_')
df_jeopardy.insert(loc=0, column='id', value=range(len(df_jeopardy)))
```

```python
display(df_jeopardy)
```

```python
def filter_data(df, list_of_words):
    filter = lambda x: all(word.lower() in x.lower() for word in list_of_words)
    return df.loc[df['question'].apply(filter)]

import re

def filter_data_imp(df, list_of_words, column):
    filter = lambda x: all(
       re.search(rf"\b{word.lower()}(?:'s)?\b", x.lower()) for word in list_of_words 
    )
    return df.loc[df[column].apply(filter)]
```

```python
df_test1 = filter_data(df_jeopardy, ['King', 'England'])
df_test2 = filter_data_imp(df_jeopardy, ['King', 'England'], 'question')
```

```python
df_jeopardy['value_float'] = df_jeopardy['value'].str.strip('$')\
    .str.replace(',', '')\
    .apply(
        lambda x: float(x) if x != 'no value' else 0
    )
```

```python
def count_unique(df):
    return df['answer'].value_counts()
```

```python
def count_occurance_timespan(year_start, year_end, word):
    df_jeopardy['air_date_conv'] = pd.to_datetime(df_jeopardy['air_date'], format='%Y-%m-%d')
    df_filtered = filter_data_imp(
        df_jeopardy[(df_jeopardy['air_date_conv'].dt.year >= year_start) & (df_jeopardy['air_date_conv'].dt.year < year_end)],\
            [word],\
            'question')
    return len(df_filtered)

occurance1990 = count_occurance_timespan(1990, 2000, 'computer')
occurance2000 = count_occurance_timespan(2000, 2010, 'Computer')
print(occurance1990)
print(occurance2000)
```
